```python
import pandas as pd
import numpy as np
import json
from os import listdir
from os.path import isfile, join
import copy
```

```python
path = '/Users/davidbautista/Documents/tesis/luti-back/server/'
```

```python
def timelineControlData(zona):
    if zona == 0:
        dat = pd.DataFrame([{'value': 0, 'zona': 'todas'.format(zona)}]*42)
        nombre = ''
    else:
        dat = pd.DataFrame([{'value': 0, 'zona': 'a{}'.format(zona)}]*42)
        nombre = '[a{},'.format(zona)
    onlyfiles = [f for f in listdir('{}docs/escenario1/'.format(path)) if isfile(join('{}docs/escenario1/'.format(path), f)) and 'accessibility ij{}'.format(nombre) in f]

    for filename in onlyfiles:
        try:
            file = pd.read_json('{0}docs/escenario1/{1}'.format(path, filename))
            dat['value'] += file['value']/len(onlyfiles)
            dat['time'] = file['time']
        except:
            continue
    # print("dat ***")
    # print(dat)

    return dat
```

```python
timelineControlData(82)
```

```python
for i in range(1, 128):
    dat = timelineControlData(i)
    dat.to_json('{0}docs/timelines/timeline_{1}.json'.format(path,i), orient='records')
```
